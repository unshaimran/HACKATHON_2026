```python
categorical_columns = ["gender",	"NationalITy",	"PlaceofBirth",	"StageID"	,"GradeID",	"SectionID",	"Topic", "Semester", "Relation",
                       "raisedhands",	"VisITedResources", "AnnouncementsView",
                       "Discussion",	"ParentAnsweringSurvey",	"ParentschoolSatisfaction",	"StudentAbsenceDays","dropout"]
```

```python
for col_name_in_list in categorical_columns :
  actual_col = col_name_in_list
  if actual_col not in data.columns:
    if actual_col.lower() in data.columns:
      actual_col = actual_col.lower()
    else:
      print(f"Column '{col_name_in_list}' not found in DataFrame. Skipping factroization")
      continue


  data[actual_col] = pd.factorize(data[actual_col])[0]
```

```python
# Example simple rule-based prediction
data['predicted_dropout'] = 0  # Default 0
# Rule: agar StudentAbsenceDays zyada hai ya parents satisfaction low hai
data.loc[(data['StudentAbsenceDays'] > 0) | (data['ParentschoolSatisfaction'] == 0), 'predicted_dropout'] = 1
```

```python
# Accuracy check
accuracy = (data['predicted_dropout'] == data['dropout']).mean()
print("Simple rule accuracy:", accuracy)
```

```python
# Risk score initialize
data['risk_score'] = 0
```

```python
# For each categorical feature, add its contribution
categorical_columns = ['gender','NationalITy','PlaceofBirth','StageID','GradeID','SectionID',
                       'Topic','Semester','Relation','ParentAnsweringSurvey','ParentschoolSatisfaction']
```

```python
for col in categorical_columns:
    data[col] = pd.factorize(data[col])[0]

# Tuned weights (try to improve accuracy)
weights = {
    'StudentAbsenceDays': 2.0,      # Absences matter a lot
    'ParentschoolSatisfaction': 1.8,
    'StageID': 0.5,
    'GradeID': 0.5,
    'Topic': 0.3,
    'Semester': 0.2,
    'Relation': 0.2
}
data['risk_score'] = 0
for col, w in weights.items():
    data['risk_score'] += data[col] / data[col].max() * w

# Adjusted thresholds
data['risk_label'] = pd.cut(data['risk_score'],
                          bins=[-0.1, 1.2, 2.5, data['risk_score'].max()],
                          labels=['Low', 'Medium', 'High'])

# Predicted dropout
data['predicted_dropout'] = data['risk_score'].apply(lambda x: 1 if x > 2.0 else 0)

# Accuracy
accuracy = (data['predicted_dropout'] == data['dropout']).mean() * 100
print("Tuned weighted rule accuracy:", round(accuracy, 2), "%")

# Top 20 high-risk students
top_high_risk = data[data['risk_label'] == 'High'].sort_values('risk_score', ascending=False).head(20)
print("\nTop 20 high-risk students:")
print(top_high_risk[['risk_score','risk_label','predicted_dropout']])

# Save predictions
data.to_csv('student_risk_predictions_tuned.csv', index=False)
print("\nPredictions saved as 'student_risk_predictions_tuned.csv'")
```
